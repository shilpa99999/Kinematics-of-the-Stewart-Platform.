# problem 1
In this activity we have written a function f1_a  which takes theta as input and calculates f(theta) and we have tested this function at -pi/4 and pi/4 and got the result closer to zero

```python
# problem 1

import math

def f_1a(theta):
    L1 = 2
    L2 = math.sqrt(2)
    L3 = math.sqrt(2)
    gamma = math.pi / 2
    x1 = 4
    x2 = 0
    y2 = 4
    p1 = math.sqrt(5)
    p2 = math.sqrt(5)
    p3 = math.sqrt(5)

    A2 = L3 * math.cos(theta) - x1
    B2 = L3 * math.sin(theta)
    A3 = L2 * ((math.cos(theta) * math.cos(gamma)) - (math.sin(theta) * math.sin(gamma))) - x2
    B3 = L2 * ((math.cos(theta) * math.sin(gamma)) + (math.sin(theta) * math.cos(gamma))) - y2

    N1 = B3 * (p2**2 - p1**2 - A2**2 - B2**2) - (B2 * (p3**2 - p1**2 - A3**2 - B3**2))
    N2 = (-A3) * (p2**2 - p1**2 - A2**2 - B2**2) + (A2 * (p3**2 - p1**2 - A3**2 - B3**2))
    D = 2 * ((A2 * B3) - (B2 * A3))

    x = N1 / D
    y = N2 / D

    out = N1**2 + N2**2 - p1**2 * D**2

    return out

# Example usage when theta = - pi/4
result = f_1a(-math.pi/4)
print(result)

# Example usage when theta = pi/4
result = f_1a(math.pi/4)
print(result)
```

# Activity 7

In the below code

For 0 poses, P2 values are: 1, 2, 3

For 2 poses, P2 values are: 4, 8, 9

For 4 poses, P2 values are: 5, 6

For 6 poses, P2 values are: 7

```python
# problem 7

import math
import matplotlib.pyplot as plt

L1 = L3 = 3
L2 = 3 * math.sqrt(2)
gamma = math.pi/4
P1 = 5
P3 = 3

x1 = 5
x2 = 0
y2 = 6




def equation(theta, P2):
    A2 = L3*math.cos(theta) - x1
    B2 = L3*math.sin(theta)
    A3 = L2*math.cos(theta + gamma) - x2
    B3 = L2*math.sin(theta + gamma) - y2

    D = 2*(A2*B3 - B2*A3)
    N1 = B3*(P2**2 - P1**2 - A2**2 - B2**2) - B2*(P3**2 - P1**2 - A3**2 - B3**2)
    N2 = -A3*(P2**2 - P1**2 - A2**2 - B2**2) + A2*(P3**2 - P1**2 - A3**2 - B3**2)
    f = (N1**2 + N2**2 - P1**2 * D**2)
    return f

def count_real_roots(f_values):
    return sum(1 for i in range(len(f_values)-1) if f_values[i] * f_values[i+1] < 0)

# Search over a range of P2 values
p2_values = list(range(1, 10))

pose_counts = {
    "0 poses": [],
    "2 poses": [],
    "4 poses": [],
    "6 poses": []
}

theta_values = [math.radians(deg) for deg in range(-180, 180)]

for P2 in p2_values:
    f_values = [equation(theta, P2) for theta in theta_values]
    root_count = count_real_roots(f_values)
    root_indices = [i for i in range(len(f_values)-1) if f_values[i] * f_values[i+1] < 0]
    if root_indices:
        theta_intervals = [(theta_values[i], theta_values[i+1]) for i in root_indices]
        print(f"For P2 = {P2}, root intervals in radians are:")
        for interval in theta_intervals:
            print(f"Theta: {interval[0]} to {interval[1]} radians")

    if root_count == 0:
        pose_counts["0 poses"].append(P2)
    elif root_count == 2:
        pose_counts["2 poses"].append(P2)
    elif root_count == 4:
        pose_counts["4 poses"].append(P2)
    elif root_count == 6:
        pose_counts["6 poses"].append(P2)

print(pose_counts)

for pose_count, p2_list in pose_counts.items():
    print(f"For {pose_count}, P2 values are: {', '.join(map(str, p2_list))}")
```
